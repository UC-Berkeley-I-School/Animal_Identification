```python
in_path = "/Users/madhuhegde/work/berkeley/W210/datasets/leopard/resize_256/"
out_path = "/Users/madhuhegde/work/berkeley/W210/datasets/leopard/classes_64/resize_256/"
#out_path = '/Users/madhuhegde/Google Drive/My Drive/datasets/leopard/classes_64/resize_256'

def create_dataset_folders(in_path=None, out_path=None, min_num_images=1):
    
    isExist = os.path.exists(out_path)
    if not isExist:
        os.makedirs(out_path+'test')
        os.makedirs(out_path+'train')
        
    labels = glob.glob(in_path+'/train/*')  
    num_examples = []    
    folders = []
    for label in labels:
        faces = glob.glob(label+'/face/*')
        if(len(faces) > min_num_images):
            copy_str = 'cp -r '+label+' '+out_path+'/train/'+label.split('/')[-1]
            os.system(copy_str)
            copy_str = 'cp -r '+in_path+'test/'+label.split('/')[-1]+' '+out_path+'test/'+label.split('/')[-1]
            os.system(copy_str)
            
            num_examples.append(len(faces))
            folders.append(label)
    print(len(num_examples))
    return folders


    
folders = create_dataset_folders(in_path, out_path, min_num_images=6)
print(folders)    
#example_gt_num = [count for count in num_examples if count>6]
#print(len(example_gt_num))
```

```python
in_path = '/Users/madhuhegde/work/berkeley/W210/datasets/leopard/resize_256/'
def find_dataset_folders(in_path=None, min_num_images=1):
    
    
    labels = glob.glob(in_path+'/train/*')  
    num_examples = []    
    folders = []
    for label in labels:
        faces = glob.glob(label+'/face/*')
        if(len(faces) < min_num_images):
            
            num_examples.append(len(faces))
            folders.append(label)
    
    return folders, num_examples

folders, num_examples = find_dataset_folders(in_path=in_path, min_num_images=7)
print(num_examples)
```
