# Intersection between test sets

Another team also worked on Identification. Here we found the matches between both sets.

```python
import pandas as pd
import re
import os
```

```python
raw_train='/content/gdrive/MyDrive/MIDS/W210/Animal_Identification/datasets/leopard/train/images'
```

```python
test2=pd.read_csv('/content/safari_sleuth_leopard_test.csv')
test2
```

```python
test2=test2.rename(columns={'Unnamed: 0':'id'})
test2
```

```python
test2['img_name']=test2.apply(lambda x: 'leop_%s_%s' % (x['id'],x['leopard_paths']), axis=1)
test2.head()
```

```python
test2['path']= test2.apply(lambda x: '%s/%s' % (raw_train,x['img_name']), axis=1)
test2.head()
```

```python
## Our dataset - test
raw_test='/content/drive/MyDrive/MIDS/W210/Animal_Identification/datasets/leopard/test/images'
# raw_dev='/content/gdrive/MyDrive/MIDS/W210/Animal_Identification/datasets/leopard/dev/images'
img_test= os.listdir(raw_test)
# img_dev= os.listdir(raw_dev)
```

```python
len(img_test)
```

```python
tmp=[]
for name in img_test:
  text=re.search('[0-9]+.jpg',name).group(0)
  text=text.lstrip("0")
  tmp.append(text)
tmp[:10]
```

```python
test2['both_testsets']=test2.apply(lambda x: True if x['stripped'] in tmp else False, axis=1)
test2.head()
```

```python
matches=test2[test2.both_testsets==True]
matches.head()
```

```python
matches_files=[]
for img in matches['stripped']:
  print(img)
  text=re.search('[0-9]+.jpg',img).group(0)
  text=text.lstrip("0")
  print('#### CHECK TEST #####')

  for name in img_test:
    # print(name)
    text2=re.search('[0-9]+.jpg',name).group(0)
    # print(text2)
    text2=text2.lstrip("0")
    # print(text, text2)

    if text2==text:
      print(f"Same files founded: {text2} and {text}")
      matches_files.append(name)
      break
matches_files
```

```python
%cd /content/
df_matches=pd.DataFrame(matches_files, columns=['img_name'])
df_matches.to_csv('test_comparison_set.csv')
```
