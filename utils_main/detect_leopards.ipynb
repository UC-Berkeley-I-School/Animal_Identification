# Using python detect_leopard:
*   *imgs* or *--source*: Path of image or folder of images to be detected
*   *folder* or *project*: Path of folder to store cropped images
*   *weights*: pt file from YOLO training with stored weights
*   *conf_thres*: confidence threshold to be utilized in YOLO inference
*   *detect*: Path of detect inference file
*   *flank_size*: Dimension for resizing flank images
*   *leop_size*: Dimension for resizing leop/full images
*   *flank_size*: Dimension for resizing flank images

General Directions: 
  Please ensure you are following the correct bash configurations for each parameter, otherwise inference pipeline will not work with arguments will not be parsed as they are supposed to. Confidence threshold needs to be **float**, while image resize dimensions need to be given as "--face-size 64 64", for example. This is due to the fact that arg parser will not recognize inputs of actual lists or tuples. 

  Also, for validation or train data, the detect pipeline needs to run separately for each.
